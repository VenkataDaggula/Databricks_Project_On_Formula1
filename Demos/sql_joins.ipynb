```python
%sql
create or replace temp view v_driver_2018
as
select race_year,driver_name,team,total_points,wins,rank
from driver_standings_table
where race_year=2018
```

```python
%sql
select *
from v_driver_2018
```

```python
%sql
create or replace temp view v_driver_2020
as
select race_year,driver_name,team,total_points,wins,rank
from driver_standings_table
where race_year=2020
```

```python
%sql
select * from v_driver_2020
```
